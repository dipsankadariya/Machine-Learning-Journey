```python
df=df.iloc[:,2:]
```

```python
df.head()
```

```python
#### train_test_split
```

```python
X=df.drop(columns='purchased')
y=df['purchased']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
### using ordinal encoder to encode the feature column
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])

##categories , we pass catogries we  want to encode
##we passes in order of lowest to highest value while encoding
## poor will get lowest, good will get highest and so on
```

```python
oe.fit(X_train)
```

```python
X_train=oe.transform(X_train)
X_test=oe.transform(X_test)
```
