```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
heart_disease = fetch_ucirepo(id=45)

# data (as pandas dataframes)
X = heart_disease.data.features
y = heart_disease.data.targets

# metadata
print(heart_disease.metadata)

# variable information
print(heart_disease.variables)
```

```python
df = pd.concat([heart_disease.data.features, heart_disease.data.targets], axis=1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer= SimpleImputer(strategy='most_frequent')
df[['ca', 'thal']] =imputer.fit_transform(df[['ca', 'thal']])
```

```python
df.isnull().sum()
```

```python
df['num'].nunique()
```

```python
print(df[['ca', 'thal']].isnull().sum())
```

```python
X=df.drop(columns=['num'])
y=df['num']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size= 0.2, random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree  import DecisionTreeClassifier
```

```python
clf1= LogisticRegression(max_iter=1000)
```

```python
clf2=DecisionTreeClassifier()
```

```python
clf1.fit(X_train,y_train)
```

```python
clf2.fit(X_train,y_train)
```

```python
y_pred1= clf1.predict(X_test)
y_pred2=clf2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy of LR:" ,accuracy_score(y_test,y_pred1))
print("Accuracy of DT:" ,accuracy_score(y_test,y_pred2))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred1))
```
