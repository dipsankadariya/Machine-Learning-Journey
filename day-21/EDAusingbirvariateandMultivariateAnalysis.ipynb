##  scatterplot(Numerical  -numerical)

```python


tips.head()
```

```python
#looking the  relationship between  totalbill  and tips


sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
#finding the  male and females
sns.scatterplot(x='total_bill',y='tip',data=tips,hue=tips['sex'])
```

```python
#finding the  male and females,and finding customers with   smoker
sns.scatterplot(x='total_bill',y='tip',data=tips,hue=tips['sex'],style=tips['smoker'])
```

```python
#finding the  male and females,and finding customers with   smoker  and   quantity    of customers
sns.scatterplot(x='total_bill',y='tip',data=tips,hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```


##  Distplot (Numerical  -Categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
#plotting  a   distplot of people  who  survived and didnt survive in  realtion to their age.
```

Most passengers were adults aged 20-40 years
Children (around age 5) had better survival chances
Working-age adults had lower survival rates
There were fewer elderly passengers overall

## HeatMap(Categorical -Categorical)

```python
titanic.head()
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Pclass')['Survived'].mean()
```

```python
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

```python
(titanic.groupby('Sex')['Survived'].mean()*100).plot(kind='bar')
```

```python
(titanic.groupby('Embarked')['Survived'].mean()*100).plot(kind='bar')
#c,Q,S are  named of towns
```
