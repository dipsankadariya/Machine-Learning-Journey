## One hot encoding using sklearn

```python
df.head()
```

```python
X= df.drop(columns='selling_price')
y=df['selling_price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe= OneHotEncoder(drop='first') ##drop=first, removes the first columsn for both category
##since we donot perform OneHotEncoder in all columns, and only perfrom on fuel and owner column
# we need to seperate these columns,  perfrom OneHotEncoder ,and then rejoin them
```

```python
ohe.fit_transform(X_train[['fuel','owner']])
```

```python
## A sparse matrix is a way to efficiently store data that mostly contains zeros.
##A sparse matrix where each row represents a car, and each column is a one-hot encoded category (e.g., fuel_Diesel, owner_First Owner, etc.).
##The matrix has 0s and 1s (or float64 values in sparse format).
```

```python
X_train_new=ohe.fit_transform(X_train[['fuel','owner']]).toarray()
```

```python
X_test_new=ohe.fit_transform(X_test[['fuel','owner']]).toarray()
```

```python
## from df, fetch the 2 columns(brand,km_driven), and add the new arraies ,X_train_new,X_test_new to the df
```
