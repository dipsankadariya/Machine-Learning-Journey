```python
import requests
import pandas as pd

df = pd.DataFrame()

for i in range(1, 107):
    response = requests.get(
        'https://api.themoviedb.org/3/tv/top_rated?api_key=603ae164163bf76a584b0dea9ff80800&language=en-US&page={}'.format(i)
    )
    temp_df = pd.DataFrame(response.json()['results'])[
        ['id', 'name', 'overview', 'first_air_date', 'vote_average', 'popularity', 'vote_count']
    ]
    df = pd.concat([df, temp_df], ignore_index=True)

# Output the final dataframe
print(df.head(10))
```
